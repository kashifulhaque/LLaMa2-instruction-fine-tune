```python
import pandas as pd
```

```python
df = pd.read_json("/content/databricks-dolly-15k.jsonl", lines = True)
```

```python
df.drop("category", inplace = True, axis = 1)
```

```python
df = df.fillna("")
```

```python
df.head()
```

```python
text_col = []
for _, row in df.iterrows():
  prompt = "Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completely the request.\n\n"
  instruction = str(row["instruction"])
  input_query = str(row["context"])
  response = str(row["response"])

  if len(input_query.strip()) == 0:
    text = prompt + "### Instruction:\n" + instruction + "\n### Response:\n" + response
  else:
    text = prompt + "### Instruction:\n" + instruction + "\n### Input:\n" + input_query + "\n### Response:\n" + response

  text_col.append(text)
```

```python
df.loc[:, "text"] = text_col
```

```python
df.head()
```

```python
df.to_csv("train.csv", index = False)
```
